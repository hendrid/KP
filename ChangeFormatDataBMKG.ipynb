```
import pandas as pd

from google.colab import files
uploaded = files.upload()

import io
df = pd.read_excel(io.BytesIO(uploaded['AWAN.xlsx']))
df.head(300)
```

```
df=df.rename(columns={df.columns[0]:'tgl',df.columns[1]:'jam'})
df
```

```
df2=df.dropna(subset=['tgl','jam'])
df2
```

```
df3=df2.drop(columns=[df2.columns[0],df2.columns[25], df2.columns[26],df2.columns[27],df2.columns[28],df2.columns[29]])
df3
```

```
df4 = pd.DataFrame(df3.values.ravel(), columns=['Column'])
df4
```

```
from openpyxl import load_workbook

path = "AWAN.xlsx"
book = load_workbook(path)
writer = pd.ExcelWriter(path, engine='openpyxl')
writer.book = book
writer.sheets = {ws.title: ws for ws in book.worksheets}

#df4.to_excel(writer, startrow=writer.sheets['Sheet1'].max_row, index = False,header= False)
df4.to_excel(writer, startrow=0, startcol=32, index = False,header= False)

writer.save()
```
